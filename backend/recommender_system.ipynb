```python
import pandas as pd
data = pd.read_csv("curated_products.csv")
data.head()
```

```python
data.Flavor.unique()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data['Flavor'])

input_flavor = 'Chocolate'
# input_name = 'Optimum Nutrition'
input_flavor_vector = vectorizer.transform([input_flavor])
# input_flavor_vector
cosine_similarities = cosine_similarity(input_flavor_vector, X)
cosine_similarities

dictionary = {}
for i, element in enumerate(cosine_similarities[0]):
    dictionary[i] = element

key_value_pairs = list(dictionary.items())

  # Sort the list of tuples by the values, in descending order.
key_value_pairs.sort(key=lambda x: x[1], reverse=True)

# Create a new dictionary from the sorted list of tuples.
sorted_dictionary = {}
for key_value_pair in key_value_pairs:
  key, value = key_value_pair
  if value != 1.0:
    sorted_dictionary[key] = value
top_indices = list(sorted_dictionary.keys())[:30]
#top indices : top similar products


top_product_names = data['Brand'].iloc[top_indices].tolist()
top_product_flavors = data['Flavor'].iloc[top_indices].tolist()
top_product_names = list(set(top_product_names))
top_product_flavors = list(set(top_product_flavors))
print(f"Recommended Supplements with {input_flavor} flavor:")
# print("Name : Flavor")
for i in range(min(len(top_product_names), len(top_product_flavors))):
  print(f'{top_product_names[i]} : {top_product_flavors[i]}')
```
